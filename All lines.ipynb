```python
val="_4"
receiver = pd.read_csv("flashes/csv/receiver_vanessa_zoom_set4.csv")
sender = pd.read_csv("flashes/csv/sender_sania_zoom_set4.csv")
start = max(receiver["Timestamp"].min(),sender["Timestamp"].min())
receiver = receiver[receiver["Timestamp"]>=start]
sender = sender[sender["Timestamp"]>=start]
receiver["Timestamp"] = receiver["Timestamp"].apply(lambda x: x-start)
sender["Timestamp"] = sender["Timestamp"].apply(lambda x: x-start)
receiver["Timestamp"] = receiver["Timestamp"].apply(lambda x: x*1000000)
sender["Timestamp"] = sender["Timestamp"].apply(lambda x: x*1000000)
df = sender[["Timestamp", "isFlash"]].merge(receiver[["Timestamp", "isFlash"]], on="Timestamp", suffixes=("_sender", "_receiver"), how="outer")
df = df.fillna(0)
df = df.sort_values("Timestamp").reset_index(drop=True)
df["isFlash_sender_cumsum"] = df["isFlash_sender"].cumsum()
df["isFlash_receiver_cumsum"] = df["isFlash_receiver"].cumsum()
_MAX_ = df['isFlash_receiver_cumsum'].max()
df['isFlash_receiver_cumsum'] = df['isFlash_receiver_cumsum']/_MAX_
df['isFlash_sender_cumsum'] = df['isFlash_sender_cumsum']/_MAX_
col_dict = dict()
for col in df.columns:
    if col=="Timestamp":
        continue
    col_dict[col] = col+val
df.rename(columns=col_dict, inplace=True)
df4 = df.copy()
```
